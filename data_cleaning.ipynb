```python
q = """
        Please tell me if this includes an academic citation. By academic citation, it doesnt necessarily 
        mean texts with explicit citations in it. Mark false if the text is subjective or contains opinionated information
        without citing valid evidence. If the text contains unarguable facts without citing its 
        source (i.e. scientific facts or common knowledge), it's still a viable citatiion. Some texts are also
        citing images (i.e. (figure 3), shown in pink, shown in blue, shown in image 1, etc.) which are also 
        citations. If so, tell me what format it is and clean the segment so that it's one coherent sentence. 
        If the text uses a footnote, clearly indicate by wrapping the indicated number or roman numeral with parentheses. If the 
        citation.xlsx number is the same, make sure it has all rows in the citation number has the same
        citation format value. 
        Text: {}
        Output JSON only in the following format:
        {{  "cleaned_sentence": str,
            "includes_citation": bool,
            "footnote": bool,
            "citation_format": str
        }}
        """
```
